```python
import pandas as pd
import numpy as np
import os

print("Libraries loaded ✓")
```

```python
df = pd.read_csv("../data/processed/all_schemes_nav.csv")

print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nFirst 5 rows:\n", df.head())
```

```python
print("Missing Values:\n")
print(df.isnull().sum())

print(f"\nTotal missing: {df.isnull().sum().sum()}")
```

```python
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

df = df.drop_duplicates()
print(f"Shape after removing duplicates: {df.shape}")
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')

df['nav'] = pd.to_numeric(df['nav'], errors='coerce')

print("Fixed Data Types:\n", df.dtypes)
print("\nSample:\n", df.head())
```

```python
print("NAV Statistics:\n")
print(df['nav'].describe())

invalid_nav = df[df['nav'] <= 0]
print(f"\nInvalid NAV rows (zero or negative): {len(invalid_nav)}")

df = df[df['nav'] > 0]
print(f"Shape after cleaning: {df.shape}")
```

```python
df = df.sort_values(['scheme_code', 'date'])
df = df.reset_index(drop=True)

print("Sorted data sample:\n")
print(df.head(10))
```

```python
os.makedirs("../data/processed", exist_ok=True)

df.to_csv("../data/processed/nav_cleaned.csv", index=False)
print(f"✓ Cleaned data saved!")
print(f"Final shape: {df.shape}")
```

```python
print("="*50)
print("DATA CLEANING SUMMARY")
print("="*50)
print(f"Total records: {df.shape[0]}")
print(f"Total columns: {df.shape[1]}")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
print(f"Schemes: {df['scheme_name'].nunique()}")
print(f"Missing values remaining: {df.isnull().sum().sum()}")
print(f"Duplicate rows: {df.duplicated().sum()}")
print("="*50)
print("✓ Data is clean and ready for analysis!")
```
